```python
import yfinance as yf
```

```python
# Fetch PayPal stock data
ticker = "PYPL"
paypal = yf.Ticker(ticker)
```

```python
# Basic stock information
info = paypal.info
print("Company Name:", info.get("longName"))
print("Sector:", info.get("sector"))
print("Current Price:", info.get("currentPrice"))
print("Market Cap:", info.get("marketCap"))

# Historical price data
hist = paypal.history(period="1y")  # Fetch past year's data
print(hist.head())

# Financials (Income Statement, Balance Sheet, Cashflow)
financials = paypal.financials
balance_sheet = paypal.balance_sheet
cashflow = paypal.cashflow

print("Income Statement:\n", financials)
print("Balance Sheet:\n", balance_sheet)
print("Cashflow:\n", cashflow)
```
